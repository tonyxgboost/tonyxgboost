```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('movies.csv')
print(data.shape)
data.head()
```

# Selecting the relevant features

```python
selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
print(selected_features)
```

```python
data.isnull().sum()
```

```python
# replacing the null values with null strings

for feature in selected_features:
    data[feature] = data[feature].fillna('')
    
    
```

```python
combine_features = data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']
print(combine_features)
```

```python
#converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combine_features)
```

```python
print(feature_vectors)
```

```python
#finding the similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# Getting the movie name from the user
movie_name = input('Enter a movie name: ')
```

```python
# Creating a list with all the movie name given in the dataset

list_of_all_titles = data['title'].tolist()
print(list_of_all_titles)
```

```python
# Finding the close match for the movie given by the user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title
index_of_the_movie = data[data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# Getting a list of similar movies
# Enumerate; keeps count of iterations

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# Sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# printing the name of the similar movies based on the index
print('movies suggested for you: \n')

i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = data[data.index == index]['title'].values[0]
    if(i<30):
        print(i, '.', title_from_index)
        i +=1
    
```
